1) Для организации аналитики по данному продукту:
1. я начала бы с постановки задачи владельцу продукта в своей компании, определив с ним цель в виде повышения потока клиентов на курс Data Science, выслушав первичные предложения и замечания. 
2. владелец продукта формирует требования и цель команде, состоящей из 
- BI аналитика
- Разработчика
- Специалиста по БД
- Data Engineer
- Специалиста DS
- Аналитика данных
- Менеджера по работе с клиентами (опосредованно относится к команде, но если потребуются дополнительные сборы информации, которых нет еще в базе, необходим для участия)

 {Примерные признаки для будущей модели:
 - возраст
 - высшее образование (сфера)
 - предыдущий опыт работы в общих чертах (10 лет в IT на позиции разработчика, 10 лет в логистике и нет опыта, только закончил инстиут и тд}
 - средний уровень дохода, исходя из опыта (если клиент не указывает при опросе, берем средний по рынку в зависимости от позиции на должности и регионе)
 - город проживания и работы
 - откуда узнал и пришел(реклама, рекомендации других учеников и тд) 
 - хочет ли сменить город проживания
 - начало обучения по курсу (дата)
 - конец обучения по курсу (дата)
 - дата получения диплома
 - фактически затраченное на обучение время (есть сутденты, которые продливают курс и судя по всему их большинство.   Если человек еще не получил диплом,но время уже продлено, тоже учитывать)
 - трудоустроился ли после обучения
 - как скоро трудоустроился
 - уровень дохода на новом месте
 - наименование компании трудоустройства
 - сфера применения DS в новой компании

 - затраты на рекламу в интернете
 - затраты на зп преподавателей и другого персонала, включая менеджеров, организаторов и курьеров
 - затраты на аренду и все интернет-ресурсы для обучения и тд
 - можно включить скидку на новый курс за развернутый ответ на вопросы и рекомендации центра обучения (особо интересует трудоустройство и рекомендации, так как дает положительные директивы на рынке)
 

3. Далее работа идет по цепочке, где-то процесс распараллеливается
 - BI аналитик занимается диагностированием текущего состояния бизнеса, смотрит на базы данных и формирует требования к дополнительным данным и общему виду, моделирует исходы событий и принимая их в учет, корректирует требования к сбору данных
 - Разработчик, Специалист БД и Data Engineer где-то в такой последовательности, а в каких-то моментах распараллеленно настраивают получение данных(разраб - при этом при нехватке каких-либо данных обращается через менеджера продукта к менеджеру по клиентам для получения недостающих), Специалист БД организует хранение и доступ к данным, Data Engineer - сбор, подготовку, структурирование выгрузок
 - Специалист DS проводит EDA, строит модели, транслирует результат метрик, если DS обнаруживает недостатки в данных, он обращается к специалистам пунктом выше и менеджеру продукта
 
4. Аналитик данных проводит исследование на значимость результатов, визуализацию результатов. 
5. К работе аналитика данных подключается Аналитик BI и приводит результаты исследования в удобоваримый для бизнеса вид
6. Проект принимает менеджер продукта, транслирует владельцу продукта, владелец продукта в свою очередь вносит коррективы/презентует решение руководству/бизнесу/заказчику

2) ключевые метрики продукта:
- Стоимость привлечения клиента
- Количество привлеченных клиентов

Основные данные в поиске инсайтов - данные обратной связи, в первую очередь статистика трудоустройства после курсов
