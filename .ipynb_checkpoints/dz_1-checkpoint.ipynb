Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
#среднее - не находится
statistics.mean(df['Freedom'])
sum(df['Freedom'])/len(df['Freedom'])

#      1) МЕДИАНА
# Извлекаем Freedom 
Freedom = df['Freedom']

# Находим их количество
num_Freedom = len(Freedom)

# Сортируем в порядке возрастания
sorted_Freedom = sorted(Freedom)

# Ищем индекс среднего элемента
# если количество элементов четное, то берем среднее двух элементов в середине
middle = (num_Freedom // 2)
if num_Freedom%2==0:
  result = (sorted_Freedom[middle-1] + sorted_Freedom[middle])/2
else:
  result = sorted_Freedom[middle]
# Находим медиану
print('Медиана: ', result)
```

2) МОДА

```python
print('Значение моды: ', statistics.mode(Freedom))
```

2й способ

```python
# Создаём пустой словарь, в котором будем считать количество появлений значений уровня свободы
Freedom_counts = {}
for f in Freedom:
    if f not in Freedom_counts:
        Freedom_counts[f] = 1
    else:
        Freedom_counts[f] += 1

# Проходимся по словарю и ищем максимальное количество повторений
# Алгоритм поиска максимума
maxf = 0
mode_Freedom = None
for k, v in Freedom_counts.items():
    if maxf < v:
        maxf = v
        mode_Freedom = k
print('Значение моды: ', mode_Freedom, 'Количество встречаемости: ', maxf)
```

Почему нельзя найти моду?

# потому что мода - это часто повторяющиеся значения, а в данной выборке все они уникальны

6) Стандартное отклонение - из-за разброса не определен

```python
#первый метод
statistics.pstdev(df['Freedom'][df['Freedom'].notnull()])
#второй
def stdev(nums):
    diffs = 0
    avg = sum(nums)/len(nums)
    for n in nums:
        diffs += (n - avg)**(2)
    return (diffs/(len(nums)-1))**(0.5)

print(stdev(df['Freedom'][df['Freedom'].notnull()]))
#третий
df['Freedom'].std()
```

Сделать частотный график колонки AveragePrice. Какое это распределение?

```python
t = df['AveragePrice'].hist(bins=30, density=True)
mu, sigma = df['AveragePrice'].mean(), df['AveragePrice'].std()
x_min, x_max = df['AveragePrice'].min(), df['AveragePrice'].max()
x = np.linspace(x_min, x_max, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (x - mu)**2 / (2 * sigma**2) ), 
         linewidth=2, color='r')
plt.xlabel('интервал выборки')
plt.ylabel('количество элементов')
plt.title('Частотный график выборки сгенерированной теоритически методом №2')
plt.grid(True)
plt.show()
```

Написать временные предсказания цен на авокадо типа organic для выбранного региона: LasVegas, начиная с августа 2017 года до конца марта 2018. В качестве тренировочных данных использовать данные до июля 2017 года включительно.
В качестве примера использовать лекционный материал в lecture_1.ipynb

```python
df['Date'] = pd.to_datetime(df['Date'])
PREDICTION_TYPE = 'organic'
df_conv = df[df.type == PREDICTION_TYPE]
regions = df_conv.groupby(df_conv.region)
PREDICTING_FOR = "LasVegas"
date_price = regions.get_group(PREDICTING_FOR)[['Date', 'AveragePrice']].reset_index(drop=True)
```

```python
date_price.plot(x='Date', y='AveragePrice', kind="line")
```

```python
date_price = date_price.rename(columns={'Date':'ds', 'AveragePrice':'y'})
date_price = date_price.sort_values(by='ds')
amount_july_17 = int(len(date_price)*0.8)
train_price = date_price[date_price['ds']<=pd.Timestamp('2017-07-31')]
test_price = date_price[date_price['ds']>=pd.Timestamp('2017-07-31')]
train_price.head()
```

```python
from fbprophet.forecaster import Prophet

m = Prophet(
    daily_seasonality=True,
    weekly_seasonality=True
)
m.fit(train_price)
```
