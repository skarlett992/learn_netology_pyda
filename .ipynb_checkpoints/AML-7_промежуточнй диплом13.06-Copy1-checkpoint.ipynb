# обработка категориальных признаков, выделение общих признаков и  таргет-переменной

```python
#закодиуем обработанные признаки и выделим таргет-переменную
y = data_X['Item_Outlet_Sales']
data = data_X.drop(['Item_Outlet_Sales', 'Name_Item_Identifier'], axis=1)
X = pd.get_dummies(data, columns=cat_feat)
```

```python
#проверим, какое количество фич у нас получилось из предварительно предобработанных категориальных и остальных
X.shape, y.shape
```

# разделение выборки на тестовую и тренировочную, построение моделей

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.shape, y_train.shape, X_test.shape, y_train.shape
```

# используем метрику r2, так как она самая простая и удобная для задач регрессии, так как коэффициент детерминации позволяет оценить улучшение предсказания моделью

```python
#Линейную регрессию
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
y_pred_test = model_lr.predict(X_test)
y_pred_train = model_lr.predict(X_train)
print("R2: \t", r2_score(y_train, y_pred_train),r2_score(y_test, y_pred_test))
```

```python
#Бустинг
model_cat = CatBoostRegressor(n_estimators=1000, learning_rate=0.2, max_depth=4, silent=True)
model_cat.fit(X_train, y_train)
y_pred_test = model_cat.predict(X_test)
y_pred_train = model_cat.predict(X_train)
print(mean_squared_error(y_train, y_pred_train), mean_squared_error(y_test, y_pred_test))
print(r2_score(y_train, y_pred_train), r2_score(y_test, y_pred_test)  )   
```

Модель не имеет особо хорошего качества. Теперь попробуем стекинг

```python
ensemble = SuperLearner()

# Build the first layer
ensemble.add([RandomForestRegressor(n_estimators=500, max_features=15), Ridge(alpha=1.0)])

# Attach the final meta estimator
ensemble.add_meta(LinearRegression())
ensemble.fit(X_train, y_train)

# Predict
preds_train, preds_test = ensemble.predict(X_train), ensemble.predict(X_test)
r2_score(y_train, preds_train), r2_score(y_test, preds_test)
```

качество модели уже лучше, но все равно не самое высокое. Применим дополнительное разбиение

```python
# Разбиваем на train/test
splitter = KFold(n_splits=10, random_state=555)
#первым в выборку идет 
for train_index, test_index in splitter.split(data_X, y):
    d_train = data_X.iloc[train_index]
    d_test = data_X.iloc[test_index]
    
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]
```

```python
# Создаем дамми-переменные для категорий
dummy_train = pd.get_dummies(d_train[cat_feat], columns=cat_feat)
dummy_test = pd.get_dummies(d_test[cat_feat], columns=cat_feat)

dummy_cols = list(set(dummy_train) & set(dummy_test))

dummy_train = dummy_train[dummy_cols]
dummy_test = dummy_test[dummy_cols]

# 
X_train = pd.concat([d_train[num_feat], dummy_train], axis=1)

X_test = pd.concat([d_test[num_feat], dummy_test], axis=1)
```

```python
num_feat = ['Item_Weight', 'Item_Visibility', 'Item_MRP']
```

```python
# Разбиваем на train/test
train_median = d_train[num_feat].median()

X_train_lin = pd.concat([d_train[num_feat].fillna(train_median), dummy_train], axis=1)

X_test_lin = pd.concat([d_test[num_feat].fillna(train_median), dummy_test], axis=1)

scaler = StandardScaler()
scaler.fit(X_train_lin[num_feat])

X_train_lin[num_feat] = scaler.transform(X_train_lin[num_feat])
X_test_lin[num_feat] = scaler.transform(X_test_lin[num_feat])
```
