f(x) = x5 + 4sin(2x) + cos(3x+3) в точке x = 1


```python
import matplotlib.pyplot as plt
import numpy as np
import pylab

from scipy.misc import derivative
from scipy.misc import electrocardiogram
from scipy.optimize import fmin, brute
from scipy.signal import find_peaks
```

```python
from scipy.misc import derivative
def f(x):
    return x**5 + (4*np.sin(2*x)) + np.cos((3*x)+3)
derivative(f, 1.0, dx=1e-6)
```

```python
def f(x):
    return 5*(x ** 4) + 8*np.cos(2*x) - 3*(np.sin(3*x)+3)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
```

```python
print(f(1))
```

```python
def f(x):
    return 20*(x ** 3) + 16*np.sin(2*x) - 9*(np.cos(3*x)+3)
```

```python
f2x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
```

```python
print(f(1))
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.plot(x, f2x, label='2derivative')
plt.legend()
plt.show()
```
