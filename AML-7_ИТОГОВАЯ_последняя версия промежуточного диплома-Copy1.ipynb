выделим категориальные признаки и обработаем их, отфильтруем непрерывные и сформируем X и y

```python
#cat_feat = list(data1.dtypes[data1.dtypes ==object].index)#.append('Outlet_Establishment_Year')
cat_feat = ['Item_Identifier', 'Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 
             'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']
y = data['Item_Outlet_Sales']
data_X = data1.drop(['Item_Outlet_Sales'], axis=1)
X = pd.get_dummies(data_X, columns=cat_feat)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=31)
#randon_state- как бьет тестовую выборку(если указан, то при каждом запуске этой ункции разбиение будет идентичным)
estimator = LinearRegression()
estimator.fit(X_train, y_train) # Обучение

y_pred_test = estimator.predict(X_test)

y_pred_train = estimator.predict(X_train)

print("R2: \t", r2_score(y_train, y_pred_train),r2_score(y_test, y_pred_test))
#доля объясненной дисперсии
print("RMSE: \t", np.sqrt(mean_squared_error(y_train, y_pred_train)), np.sqrt(mean_squared_error(y_test, y_pred_test)))
#средняя квадратичная ошибка(чем больше, тем хуже)
print("MAE: \t", mean_absolute_error(y_train, y_pred_train), mean_absolute_error(y_test, y_pred_test))
#средняя абсолютная ошибка
```

Кросс-валидация и построение модели леса

```python
skf = KFold(n_splits=10, random_state=None, shuffle=False)
train_metric, test_metric = [], []
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    model = RandomForestRegressor(n_estimators=500, max_features=5)
    model.fit(X_train, y_train)
    y_pred_train, y_pred_test = model.predict(X_train), model.predict(X_test)
    mean_squared_error(y_train, y_pred_train), mean_squared_error(y_test, y_pred_test)
    train_metric.append(r2_score(y_train, y_pred_train))
    test_metric.append(r2_score(y_test, y_pred_test))
    print(r2_score(y_train, y_pred_train), r2_score(y_test, y_pred_test))
print(sum(train_metric)/len(train_metric))
print(sum(test_metric)/len(test_metric))
```
