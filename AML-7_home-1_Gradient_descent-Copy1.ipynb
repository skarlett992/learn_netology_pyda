```python
import numpy as np
import random
from sklearn.datasets.samples_generator import make_regression 
from sklearn.datasets.samples_generator import make_classification
import pylab 
from matplotlib.pyplot import scatter, plot
from scipy import stats
```

```python
x = np.c_[ np.ones(m), x]
x[:5]
```

```python
alpha=0.01
def gradient_descent_classic(alpha, x, y, numIterations):
    """
    Функция реализует алгоритм градиентного спуска. На каждом шаге выводится значение функции потерь
    """
    #берем количество элементов в выборке(чтобы потом нанего делить в формуле)
    def sigma(x):
        return 1/(1+np.exp(-x))

    m = x.shape[0] # 100
    theta = np.ones(3) # [ 1.  1.] ВЕСА МОДЕЛИ!!! -  k и b делаем равным 1 и 1 и затем они будут у нас меняться
    x_transpose = x.transpose() # транспонированная матрица x
    
    for iter in range( 0, numIterations ):
        #hypothesis =y_pred
        hypothesis = sigma(np.dot(x,theta)) # матричное произведение
        
        J = np.sum(-y*np.log(hypothesis)-(1-y)*np.log(1-hypothesis))  # функция потерь
        
#         print( "iter %s | J: %.3f" % (iter, J) )
        
        gradient = np.dot((hypothesis-y), x) / m         
        theta = theta - alpha * gradient  # update
    
    #theta - искомое значение k и b (kx+b) в регрессии - градиент конечный
    return theta
```

```python
%%time
theta = gradient_descent_classic(alpha, x, y, 10000)
```
